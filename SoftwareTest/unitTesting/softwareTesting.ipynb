```python
# 1. Write a program to test a function that checks if a number is even or odd.

import unittest

# Function to check even or odd
def is_even(num):
    return num % 2 == 0


class TestEvenOdd(unittest.TestCase):

    def test_even(self):
        self.assertTrue(is_even(10))

    def test_odd(self):
        self.assertFalse(is_even(9))


# Run tests in Google Colab
unittest.main(argv=[''], exit=False)
```

```python
# 2. Write unit tests for a function that finds the maximum of two numbers.

import unittest

def max_of_two(a, b):
    return a if a > b else b

class TestMaxOfTwo(unittest.TestCase):

    def test_first_bigger(self):
        self.assertEqual(max_of_two(10, 5), 10)

    def test_second_bigger(self):
        self.assertEqual(max_of_two(3, 8), 8)

    def test_both_equal(self):
        self.assertEqual(max_of_two(7, 7), 7)

unittest.main(argv=[''], exit=False)

```

```python
# 3. Test a function that counts the number of vowels in a string

import unittest

def count_vowels(s):
    vowels = "aeiouAEIOU"
    return sum(1 for ch in s if ch in vowels)

class TestCountVowels(unittest.TestCase):

    def test_simple_word(self):
        self.assertEqual(count_vowels("apple"), 2)

    def test_no_vowels(self):
        self.assertEqual(count_vowels("rhythm"), 0)

    def test_mixed_case(self):
        self.assertEqual(count_vowels("EduCaTiOn"), 5)

    def test_empty_string(self):
        self.assertEqual(count_vowels(""), 0)

unittest.main(argv=[''], exit=False)
```

```python
# 4. Test a function that finds the second largest number in a list.

import unittest

def second_largest(nums):
    nums = list(set(nums))     # remove duplicates
    nums.sort()
    return nums[-2]            # second largest

class TestSecondLargest(unittest.TestCase):

    def test_normal_list(self):
        self.assertEqual(second_largest([10, 20, 30, 40]), 30)

    def test_with_duplicates(self):
        self.assertEqual(second_largest([5, 1, 5, 3]), 3)

    def test_negative_numbers(self):
        self.assertEqual(second_largest([-1, -5, -3, -2]), -2)

    def test_small_list(self):
        self.assertEqual(second_largest([2, 1]), 1)

unittest.main(argv=[''], exit=False)
```

```python
# 5. Write unit tests for a function that returns the sum of all even numbers in a list.

import unittest

def sum_even(nums):
    return sum(n for n in nums if n % 2 == 0)

class TestSumEven(unittest.TestCase):

    def test_normal_list(self):
        self.assertEqual(sum_even([1, 2, 3, 4, 5, 6]), 12)

    def test_no_even_numbers(self):
        self.assertEqual(sum_even([1, 3, 5]), 0)

    def test_all_even_numbers(self):
        self.assertEqual(sum_even([2, 4, 6, 8]), 20)

    def test_negative_numbers(self):
        self.assertEqual(sum_even([-2, -4, 3, 5]), -6)

unittest.main(argv=[''], exit=False)
```

```python
# 6. Write a function that calculates the grade of a student based on marks and test it.

import unittest

def get_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    elif marks >= 60:
        return "D"
    else:
        return "F"

class TestGetGrade(unittest.TestCase):

    def test_grade_A(self):
        self.assertEqual(get_grade(95), "A")

    def test_grade_B(self):
        self.assertEqual(get_grade(85), "B")

    def test_grade_C(self):
        self.assertEqual(get_grade(75), "C")

    def test_grade_D(self):
        self.assertEqual(get_grade(65), "D")

    def test_grade_F(self):
        self.assertEqual(get_grade(40), "F")

unittest.main(argv=[''], exit=False)
```
